## Recursion 

In python , a function can call other functions.It is also possible for function to call itself . This is called recursion 

### Example 1 :

```python
# print factorial of a function using recursion 
def factorial(n):
    '''
    recursive function to find factorial of a number
    '''
    return 1 if n==1 else (n*factorial(n-1))
```

```python
factorial(5)
```

### Advantages 
1. Recursive function make the code look clean and elegant 
2. Complex task can be broken down into simpler sub-problems 
3. sequence generation is easier with recursion than nested iteration

### Disadvantages 
1. Takes up more memory and time 
2. Hard to debug
3. Logic can be harder sometimes 



### Example 2 :


```python
## Generate fibonacci sequence 

def fibonacci(num):
    '''
    recursive function to print fibonacci sequence
    '''
    return num if num<=1 else fibonacci(num-1)+fibonacci(num-2)

```

```python

fibonacci(0)
```

```python
fibonacci(1)
```

```python
fibonacci(3)
```

```python
fibonacci(4)
```

```python
fibonacci(5)
```

```python
# 1st 10 fibonacci sequence

for i in range(0,11):
    print(fibonacci(i))
```

## Python built-in Exceptions


```python
dir(__builtins__)
```

Errors are handled gracefully in except block 

```python
import sys
lst = ['b',0,2]
for entry in lst:
    try:
        print("The entry is ",entry)
        r = 1/int(entry)
        
    except:
        print("oops!", sys.exc_info()[0], " occured.")
        print("***********")
    
print(f"the reciprocal of {entry} is {r}")    
```

```python
lst = ['b',0,2]
for entry in lst:
    try:
        print("The entry is ",entry)
        r = 1/int(entry)
        
    except Exception as e:
        print("oops!",e, " occured.")
        print("***********")
    
print(f"the reciprocal of {entry} is {r}")   
```

```python
## SEPERATE EXCEPT BLOCK for each error

lst = ['b',0,2]
for entry in lst:
    try:
        print("The entry is ",entry)
        r = 1/int(entry)
        
    except(ValueError):
        print("value error occured")
        print("***********")
    except(ZeroDivisionError):
        print("ZeroDivisionError occured ")
        print("***********")
    except:
        print("some other error")
print(f"the reciprocal of {entry} is {r}")   
```

In the above case the error has been raised by the interpreter. Programatically we can also raise our own error

```python
try:
    num = int(input("Enter a positive number"))
    if num<=0:
        raise ValueError("Error: Entered non positive number")
except ValueError as e:
    print(e)
        
```

## Working with multiple figures and axis

```python
def f(t):
    return np.exp(t)*np.cos(2*np.pi*t)
t1= np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,0.5,0.02)

plt.figure(1) # first figure
plt.subplot(211) # first subplot in first figure
plt.grid()
plt.plot(t1, f(t1), 'b-')

plt.subplot(212)  # 2nd subplot in first figure
plt.plot(t2,np.cos(2*np.pi*t2), 'r--')
plt.show()
```

## Concatenate Data Frames

```python
india_weather = pd.DataFrame({
    "city" : ['mumbai', 'delhi', 'bangalore'],
    "temperature":[32, 45, 30],
    "humidity":[80,60,78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city" : ['new york', 'pittsburg', 'chicago'],
    "temperature":[21,14,35],
    "humidity":[68,65,75]
})
us_weather
```

```python
## concate 2 dataframes
df = pd.concat([india_weather, us_weather])  #common columns
df
```

```python
# for continuous index
df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

```python
df = pd.concat([india_weather, us_weather], axis=1)
```

```python
df
```

## Merging Dataframes

```python
temperature_df =  pd.DataFrame({
    "city": ['mumbai', 'delhi', 'bangalore', 'hyderabad'],
    "temperature":[35,45,30,40],
    
})
temperature_df
```

```python
humidity_df = pd.DataFrame({
    "city":["mumbai", "delhi", "bangalore"],
    "humidity":[68,65,75]
})
humidity_df
```

```python
## merge 2 dataframes
df = pd.merge(temperature_df, humidity_df, on="city") # it drops hyderabad 
df
```

```python
## OUTER-Join

df1 = pd.merge(temperature_df, humidity_df, on="city", how='outer')
df1
```
